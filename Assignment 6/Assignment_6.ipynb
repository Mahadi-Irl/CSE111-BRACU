***# Write your code here***

```python
# import the datetime library to access the current year

import datetime

# create class

class Employee:

    # define the constructor method
    def __init__(self, name, workingPeriod):
        self.name = name
        self.workingPeriod = workingPeriod
    
    # define a class method to create class instances accordingly
    @classmethod
    def employeeByJoiningYear(cls, name, year):
        workingPeriod = datetime.datetime.today().year - year
        return Employee(name, workingPeriod)
    
    # define a static method to check if an employee is experienced or not
    @staticmethod
    def experienceCheck(working_Period, gender):
        if working_Period < 3:
            experience  = 'not experienced'
        else:
            experience  = 'experienced'
        if gender.lower() == 'male':
            pronoun = 'He'
        elif gender.lower() == 'female':
            pronoun = 'She'
        return f'{pronoun} is {experience}'


# tester code

employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))

# end
```

```python
# import the math library to access the value of pi

import math

# create class

class Cylinder:

    # take two class variables to hold the radius and height
    radius = 5
    height = 18
    
    # define the constructor method
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        print(f'Default radius={Cylinder.radius} and height={Cylinder.height}.')
        Cylinder.radius = radius
        Cylinder.height = height
        print(f'Updated: radius={radius} and height={height}.')
    
    # define a class method to create class instances accordingly
    @classmethod
    def swap(cls, height, radius):
        return Cylinder(radius, height)
    
    # define a class method to create class instances accordingly
    @classmethod
    def changeFormat(cls, details):
        temp_list = details.split('-')
        radius = float(temp_list[0])
        height = float(temp_list[1])
        return Cylinder(radius, height)
    
    # define a static method to get the area of the cylinder
    @staticmethod
    def area(radius, height):
        area = 2 * math.pi * radius * (radius + height)
        print(f'Area: {area}')
    
    # define a class method to get the volume of the cylinder
    @staticmethod
    def volume(radius, height):
        volume = math.pi * (radius ** 2) * height
        print(f'Volume: {volume}')


# tester code

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)

# end
```

‏‏‎ ‎‏‏‎ ‎‏‏‎ ‎‏‏‎ ‎‏‏‎ ‎‏‏‎ ‎‏‏‎ ‎‏‏‎ ‎i. Branch_sell = quantity * 400\
5\. Calculation of branch’s sell percentage = (branch’s sell / total sell) * 100

```python
# create class

class SultansDine:

    # take three class variables to hold total number of branches, total sell of branches and references of all the brances
    total_branch = 0
    total_sell = 0
    branch_list = []

    # define the constructor method
    def __init__(self, branch):
        self.branch = branch
        SultansDine.total_branch += 1
        SultansDine.branch_list.append(self)

    # define an instance method to get sell amount from quantity
    def sellQuantity(self, quantity):
        if quantity < 10:
            self.sell = quantity * 300
        elif quantity < 20:
            self.sell = quantity * 350
        else:
            self.sell = quantity * 400
        SultansDine.total_sell += self.sell
    
    # define a class method to get class variable values in the required format
    @classmethod
    def details(cls):
        print(f'Total Number of branch(s): {cls.total_branch}')
        print(f'Total Sell: {cls.total_sell} Taka')
        if len(SultansDine.branch_list) != 0:
            for i in SultansDine.branch_list:
                print(f'Branch Name: {i.branch}, Branch Sell: {i.sell} Taka')
                print(f'Branch consists of total sell\'s: {round((i.sell / SultansDine.total_sell) * 100, 2)}%')

    # define an instance method to get instance variable values in the required format
    def branchInformation(self):
        print(f'Branch Name: {self.branch}')
        print(f'Branch Sell: {self.sell} Taka')

# tester code

SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()

# end
```
