# **Task - 1**

```python
# create PlayerEarning class
class PlayerEarning:
    
    def __init__(self, name):
        self.name = name
    
    # define a function to calculate the bonus
    def calculateTotal(self, earning, goals = None):
        self.earning = earning
        if goals is not None:
            if goals > 30:
                self.bonus = int((5/100) * earning + 10000)
            else:
                self.bonus = int((5/100) * earning)
        else:
            self.bonus = 0
    
    # define a function to print details accordingly
    def printDetails(self):
        print(f'Player Name: {self.name}')
        print(f'Player Season Earning without bonus: {self.earning}')
        print(f'Bonus: {self.bonus}')
        print(f'Player Season Earning After Bonus: {self.earning + self.bonus}')


#---------------------------------------------------------------------------------------------------------------------------------

print("**********************")
player1 = PlayerEarning('Buffon')
player1.calculateTotal(250000)
player1.printDetails()
print("\n**********************")
player2 = PlayerEarning('Dybala')
player2.calculateTotal(250000, 31)
player2.printDetails()
print("\n**********************")
player3 = PlayerEarning('Cuadrado')
player3.calculateTotal(250000, 20)
player3.printDetails()
```

```python
# create myList class
class myList:
    
    def __init__(self, *num):
        self.num = num
    
    # define a function to calculate the sum of all the items in the list
    def sum(self):
        summ = sum(self.num)
        print(f'Sum: {summ}')
    
    # define a function to merge a new list to the old list
    def merge(self, *num):
        self.num += num
    
    # define a function to calculate the average of the items in the list
    def average(self):
        if len(self.num) == 0:
            avrg = 0
        else:
            avrg = sum(self.num) / len(self.num)
        print(f'Average: {avrg}')


#---------------------------------------------------------------------------------------------------------------------------------

l1 = myList(2,3,4,5,6)
l1.sum()
l1.merge(4,5,9)
l1.sum()
l1.average()
print("-----------------------------")
l2 = myList()
l2.average()
l2.merge(1,2,4,8)
l2.sum()
```

```python
class Student:
    def __init__(self,name,ID):
        self.name = name
        self.ID = ID
    def Details(self):
        return "Name: "+self.name+"\n"+"ID: "+self.ID+"\n"

#---------------------------------------------------------------------------------------------------------------------------------

# create CSEStudent sub class
class CSEStudent(Student):
    
    def __init__(self, name, id, semester):
        self.name = name
        self.id = id
        self.semester = semester
        self.course_marks = {}
    
    def Details(self):
        return f'Name: {self.name}\nID: {self.id}\nCurrent semester: {self.semester}'
    
    def grading(self, mark):
        if mark >=85:
            return 4.0
        elif 80<=mark<=84:
            return 3.3
        elif 70<=mark<=79:
            return 3.0
        elif 65<=mark<=69:
            return 2.3
        elif 57<=mark<=64:
            return 2.0
        elif 55<=mark<=56:
            return 1.3
        elif 50<=mark<=54:
            return 1.0; 
        elif mark<50:
            return 0.0

    def addCourseWithMarks(self, *course_marks):
        for i in range(0,len(course_marks),2):
            self.course_marks[course_marks[i]] = course_marks[i+1]
    
    def showGPA(self):
        courses = len(self.course_marks)
        total_gpa = 0
        print(f'Bob has taken {courses} courses.')
        for i, j in self.course_marks.items():
            gpa = self.grading(j)
            print(f'{i}: {gpa}')
            total_gpa += gpa
        cgpa = total_gpa / courses
        print(f'GPA of Bob is: {round(cgpa, 2)}')

#---------------------------------------------------------------------------------------------------------------------------------

Bob = CSEStudent("Bob","20301018","Fall 2020")
Carol = CSEStudent("Carol","16301814","Fall 2020")
Anny = CSEStudent("Anny","18201234","Fall 2020")
print("#########################")
print(Bob.Details())
print("#########################")
print(Carol.Details())
print("#########################")
print(Anny.Details())
print("#########################")
Bob.addCourseWithMarks("CSE111",83.5,"CSE230",73.0,"CSE260",92.5)
Carol.addCourseWithMarks("CSE470",62.5,"CSE422",69.0,"CSE460",76.5,"CSE461",87.0)
Anny.addCourseWithMarks("CSE340",45.5,"CSE321",95.0,"CSE370",91.0)
print("----------------------------")
Bob.showGPA()
print("----------------------------")
Carol.showGPA()
print("----------------------------")
Anny.showGPA()
```

```python
class Transport:
    total_traveller = 0
    def __init__(self, name, fare):
        self.name = name
        self.baseFare = fare
    def __str__(self):
        s = "Name: "+self.name+", Base fare: "+str(self.baseFare)
        return s

#---------------------------------------------------------------------------------------------------------------------------------

# create Bus and Train classes

class Bus(Transport):
    
    total_passengers = 0
    
    def __init__(self, name, fare):
        super().__init__(name, fare)
        self.passengers = {}
        print(f'Base-fare of {self.name} is {self.baseFare} Taka')
    
    def fare_counter(self, bags):
        if bags <= 2:
            fare = 0
        elif 3 <= bags <= 5:
            fare = 60
        elif bags > 5:
            fare = 105
        return self.baseFare + fare
    
    def addPassengerWithBags(self, *passenger_bags):
        for i in range(0,len(passenger_bags),2):
            self.passengers[passenger_bags[i]] = self.fare_counter(passenger_bags[i+1])
            self.total_passengers += 1
            Transport.total_traveller += 1
    
    def __str__(self):
        temp = super().__str__()
        temp += f'\nTotal Passenger(s): {self.total_passengers}\nPassenger details:'
        for i, j in self.passengers.items():
            temp+= f'\nName: {i}, Fare: {j}'
        return temp

class Train(Transport):
    
    total_passengers = 0
    
    def __init__(self, name, fare):
        super().__init__(name, fare)
        self.passengers = {}
        print(f'Base-fare of {self.name} is {self.baseFare} Taka')
    
    def fare_counter(self, bags):
        if bags <= 2:
            fare = 0
        elif 3 <= bags <= 5:
            fare = 60
        elif bags > 5:
            fare = 105
        return self.baseFare + fare
    
    def addPassengerWithBags(self, *passenger_bags):
        for i in range(0,len(passenger_bags),2):
            self.passengers[passenger_bags[i]] = self.fare_counter(passenger_bags[i+1])
            self.total_passengers += 1
            Transport.total_traveller += 1
    
    def __str__(self):
        temp = super().__str__()
        temp += f'\nTotal Passenger(s): {self.total_passengers}\nPassenger details:'
        for i, j in self.passengers.items():
            temp+= f'\nName: {i}, Fare: {j}'
        return temp

#---------------------------------------------------------------------------------------------------------------------------------

t1 = Bus("Volvo", 950)
print("=================================")
t1.addPassengerWithBags("David", 6, "Mike", 1, "Carol", 3)
print("=================================")
print(t1)
print("=================================")
t2 = Train("Silk City", 850)
print("=================================")
t2.addPassengerWithBags("Bob", 2, "Simon", 4)
print("=================================")
print(t2)
print("=================================")
print("Total Passengers in Transport: ", Transport.total_traveller )
```
