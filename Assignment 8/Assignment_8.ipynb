Write the **ComplexNumber** class so that the following code generates the output below.

---

***OUTPUT:***

RealPart: 1.0\
ImaginaryPart: 1.0\
\--------------------\
RealPart: 5.0\
ImaginaryPart: 7.0

```python
# given class

class RealNumber:
    def __init__(self, r=0):
        self.__realValue = r
    def getRealValue(self):
        return self.__realValue
    def setRealValue(self, r):
        self.__realValue = r
    def __str__(self):
        return 'RealPart: '+str(self.getRealValue())

# create child class

class ComplexNumber(RealNumber):

    def __init__(self, real = 1, imag = 1):
        super().__init__(float(real))
        self.imag = float(imag)
    
    def __str__(self):
        temp = super().__str__()
        temp += f'\nImaginaryPart: {self.imag}'
        return temp

# tester code

cn1 = ComplexNumber()
print(cn1)
print('---------')
cn2 = ComplexNumber(5,7)
print(cn2)

# end
```

# **Task - 2**
Write the **ComplexNumber** class so that the following code generates the output below.

---

***OUTPUT:***

8\
2 + 1i\
3 + 5i\
5 + 6i\
-1 - 4i

```python
# given class

class RealNumber:
    def __init__(self, number=0):
        self.number = number
    def __add__(self, anotherRealNumber):
        return self.number + anotherRealNumber.number
    def __sub__(self, anotherRealNumber):
        return self.number - anotherRealNumber.number
    def __str__(self):
        return str(self.number)

# create child class

class ComplexNumber(RealNumber):

    def __init__(self, real, imag):
        if type(real) is not int:
            real = real.number
        super().__init__(real)
        self.imag = imag
    
    def __add__(self, other):
        temp_real = super().__add__(other)
        temp_imag = self.imag + other.imag
        return ComplexNumber(temp_real, temp_imag)
    
    def __sub__(self, other):
        temp_real = super().__sub__(other)
        temp_imag = self.imag - other.imag
        return ComplexNumber(temp_real, temp_imag)
    
    def __str__(self):
        if self.imag < 0:
            temp = f'{self.number} - {-1 * self.imag}i'
        else:
            temp = f'{self.number} + {self.imag}i'
        return temp

# tester code

r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1+r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)

# end
```

```python
# given class

class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60
    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"

# create child class

class ScienceExam(Exam):

    def __init__(self, marks, time, *parts):
        super().__init__(marks)
        self.time = time
        self.parts = parts
    
    def __str__(self):
        temp = f'Marks: {self.marks} Time: {self.time} minutes '
        temp += f'Number of Parts: {len(self.parts) + 2}'
        return temp
    
    def examSyllabus(self):
        temp = super().examSyllabus()
        temp += ' , ' + (' , ').join(self.parts)
        return temp
    
    def examParts(self):
        temp = super().examParts()
        for i in range(len(self.parts)):
            temp += f'Part {i + 3} - {self.parts[i]}\n'
        return temp[:-1]

# tester code

engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())

# end
```

# **Task - 6**
Given the following class, write the code for the **Sphere** and the **Cylinder** class so that the following output is printed.

---

***OUTPUT:***

Shape name: Sphere, Area Formula: 4 * pi * r * r\
\--------------------------------------------------------------------\
Radius: 5, Height: No need\
Area: 314.159\
\==================================\
Shape name: Cylinder, Area Formula: 2 * pi * r * (r + h)\
\--------------------------------------------------------------------\
Radius: 5, Height: 10\
Area: 471.2385

```python
# given class

class Shape3D:
    pi = 3.14159
    def __init__(self, name = 'Default', radius = 0):
        self._area = 0
        self._name = name
        self._height = 'No need'
        self._radius = radius
    def calc_surface_area(self):
        return 2 * Shape3D.pi * self._radius
    def __str__(self):
        return "Radius: "+str(self._radius)

# create child classes

class Sphere(Shape3D):

    def __init__(self, name, radius):
        super().__init__(name, radius)
        print(f'Shape name: Sphere, Area Formula: 4 * pi * r * r')
    
    def calc_surface_area(self):
        self.area = super().calc_surface_area() * 2 * self._radius
    
    def __str__(self):
        temp = super().__str__()
        temp += f' Height: {self._height}'
        temp += f'\nArea: {self.area}'
        return temp

class Cylinder(Shape3D):

    def __init__(self, name, radius, height):
        super().__init__(name, radius)
        self._height = height
        print(f'Shape name: Cylinder, Area Formula: 2 * pi * r * (r + h)')
    
    def calc_surface_area(self):
        self.area = super().calc_surface_area() * (self._radius + self._height)
    
    def __str__(self):
        temp = super().__str__()
        temp += f' Height: {self._height}'
        temp += f'\nArea: {self.area}'
        return temp

# tester code

sph = Sphere('Sphere', 5)
print('----------------------------------')
sph.calc_surface_area()
print(sph)
print('==================================')
cyl = Cylinder('Cylinder', 5, 10)
print('----------------------------------')
cyl.calc_surface_area()
print(cyl)

# end
```
