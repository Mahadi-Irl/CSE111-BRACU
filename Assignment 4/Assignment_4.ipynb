# **Question - 1**
Suppose your little sibling wants your help to check his math homework. He is done with his homework but wants you to see if all his results are correct. Since the student with all correct results gets 3 stars. However, you want your brother to check this on his own. So, you design a calculator for him in python. You could have given your scientific calculator but you wanted to give him a basic calculator and also wanted to see if you can even design one.

**Subtasks:**
1. Create a class called Calculator.
2. Your class shall have 1 constructor and 4 methods, namely add, subtract,
multiply and divide.
3. Now, create an object of your class. After creating an object, it should print "Let's Calculate!"
4. Then take 3 inputs from the user: first value, operator, second value
5. Now based on the given operator, call the required method and print the result.

---

**Sample Input:**

1
\
+
\
2

---

**Sample Output:**

Let's Calculate!
\
Value 1: 1
\
Operator: +
\
Value 2: 2
\
Result: 3

```python
# create class

class Calculator:

    def __init__(self):
        print('Let\'s Calculate!')

    def add(self, val_1, val_2):
        print('Value 1:', val_1)
        print('Operator: +')
        print('Value 2:', val_2)
        print('Result:', val_1 + val_2)

    def subtract(self, val_1, val_2):
        print('Value 1:', val_1)
        print('Operator: -')
        print('Value 2:', val_2)
        print('Result:', val_1 - val_2)

    def multiply(self, val_1, val_2):
        print('Value 1:', val_1)
        print('Operator: *')
        print('Value 2:', val_2)
        print('Result:', val_1 * val_2)

    def devide(self, val_1, val_2):
        print('Value 1:', val_1)
        print('Operator: /')
        print('Value 2:', val_2)
        print('Result:', val_1 / val_2)


# driver code

# create an object

obj_1 = Calculator()

# take input from the user

val_1 = int(input('Please enter the first value : '))
operator = input('Please enter the operator : ')
val_2 = int(input('Please enter the second value : '))

# call the required method based on the operator

if operator == '+':
    obj_1.add(val_1, val_2)
elif operator == '-':
    obj_1.sub(val_1, val_2)
elif operator == '*':
    obj_1.multiply(val_1, val_2)
elif operator == '/':
    obj_1.devide(val_1, val_2)

# end
```

```python
# create class

class Customer:

    def __init__(self, name):
        self.name = name

    def greet(self, name = None):
        if name == None:
            print('Hello!')
        else:
            print(f'Hello {name}!')
            
    def purchase(self, *items):
        print(f'{self.name}, you purchased {len(items)} item(s):')
        for i in items:
            print(i)


# driver code

customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")

# end
```

Design a "**Vehicle**" class. A vehicle assumes that the whole world is a 2-dimensional graph paper. It maintains its x and y coordinates (both are integers). Any new object created of the Vehicle class will always start at the coordinates (0,0).

It must have methods to move up, down, left, right and a print_position() method for printing the current coordinate.

Note: All moves are 1 step. That means a single call to any move method changes the value of either x or y or both by 1.

**\[You are not allowed to change the code below]**

---

***\# Write your class here***

\

\

---

\

\

```python
# create class

class Vehicle:

    def __init__(self):
        self.x_position = 0
        self.y_position = 0
    
    def moveUp(self):
        self.y_position += 1
    
    def moveDown(self):
        self.y_position -+ 1
    
    def moveLeft(self):
        self.x_position -= 1
    
    def moveRight(self):
        self.x_position += 1
    
    def print_position(self):
        print(f'({self.x_position},{self.y_position})')


# driver code

car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()

# end
```

```python
# create class

class Vaccine:
    
    def __init__(self, name, country, days):
        self.name = name
        self.country = country
        self.days = days


class Person:
    
    def __init__(self, name, age, occupation = 'General Citizen'):
        self.name = name
        self.age = age
        self.occupation = occupation
        self.dose = 0
        self.first_dose = 'Not Given'
        self.second_dose = 'Not Given'
    
    def pushVaccine(self, vaccine, dose = '1st Dose'):
        if self.age <= 25 and self.occupation != 'Student':
            print(f'Sorry {self.name}, Minimum age for taking vaccines is 25 years now.')
        else:
            if self.dose < 2:
                if dose == '1st Dose':
                    self.vaccine_name = vaccine.name
                    self.first_dose = 'Given'
                    self.second_dose = f'Please come after {vaccine.days} days'
                    self.dose += 1
                    print(f'1st dose done for {self.name}')
                else:
                    if self.vaccine_name == vaccine.name:
                        self.second_dose = 'Given'
                        self.dose += 1
                        print(f'2nd dose done for {self.name}')
                    else :
                        print(f'Sorry {self.name}, you can’t take 2 different vaccines')
            else:
                print(f'{self.name} has already takes 2 doses')
    
    def showDetail(self):
        print(f'Name: {self.name} Age: {self.age} Type: {self.occupation}')
        print(f'Vaccine name: {self.vaccine_name}')
        print(f'1st dose: {self.first_dose}')
        print(f'2nd dose: {self.second_dose}')


# driver code

astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")

# end
```
